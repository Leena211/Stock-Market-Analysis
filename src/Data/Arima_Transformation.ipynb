```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import os
```

```python
# Load transformed dataset
df = pd.read_csv("Transformed.csv")
df
```

```python
# Stocks mapping
print("🔄 Defining stock ticker mappings...")
stocks = {
    "Stocks_Apple": "Apple",
    "Stocks_GeneralElectric": "GeneralElectric",
    "Stocks_IBM": "IBM",
    "Stocks_Johnson&Johnson": "JohnsonJohnson",
    "Stocks_Microsoft": "Microsoft"
}
print("✅ Stocks mapping ready.\n")

# Create output directory
print("📁 Creating output directory for ARIMA processed data...")
os.makedirs("../data/ARIMA", exist_ok=True)
print("✅ Output directory '../data/ARIMA' is ready.\n")
```

```python
# Loop through each stock column
for col, name in stocks.items():
    stock_df = df[df[col] == 1][['Close']].copy()

    # ADF Test
    result = adfuller(stock_df['Close'])
    print(f"{name} ADF p-value: {result[1]}")

    if result[1] > 0.05:
        # Not stationary, apply differencing
        stock_df['Close'] = stock_df['Close'].diff().dropna()
        stock_df.dropna(inplace=True)
        print(f"{name} was differenced for stationarity")

    # Save processed file
    stock_df.to_csv(f"../data/ARIMA/{name}.csv")
```

```python
# Create output folder
output_dir = "data/ARIMA"
os.makedirs(output_dir, exist_ok=True)
print(f"📁 Output directory '{output_dir}' is ready.\n")

# List of stock columns
stock_columns = ['Stocks_Apple', 'Stocks_GeneralElectric', 'Stocks_IBM', 'Stocks_Johnson&Johnson', 'Stocks_Microsoft']
```

```python
# Save differenced data
for stock in stock_columns:
    print(f"🔍 Processing {stock}...")

    series = df[stock]
    p_value = adfuller(series)[1]
    print(f"   - ADF Test p-value: {p_value:.4f}")

    # Apply differencing if not stationary
    if p_value > 0.05:
        differenced_series = series.diff().dropna()
        print("   🔁 Series is non-stationary. Differencing applied.")
        status = "Differenced"
    else:
        differenced_series = series
        print("   ✅ Series is stationary. No differencing needed.")
        status = "Original"
```

- Checked **stationarity** using the Augmented Dickey-Fuller (ADF) test
- Applied **first-order differencing** to convert non-stationary time series to stationary
- Saved transformed (stationary) data for each stock as CSV files
- Logged and printed ADF p-values to confirm transformation effectiveness
