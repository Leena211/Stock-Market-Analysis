```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np

# Input/output paths
input_dir = "../Data/ARIMA"
output_dir = "results/ARIMA"
os.makedirs(output_dir, exist_ok=True)
print(f"📁 Output directory '{output_dir}' is ready.\n")

# Stocks list (clean names match CSV files)
stocks = [f.replace(".csv", "") for f in os.listdir(input_dir) if f.endswith(".csv")]

# Loop through each stock
for stock in stocks:
    print(f"\n🔄 Processing ARIMA for {stock}...")

    # Load data
    df = pd.read_csv(f"{input_dir}/{stock}.csv", parse_dates=['Date'], index_col='Date')
    series = df.iloc[:, 0]  # get the only column (stock price)
    series

    # Plot ACF & PACF
    print("📊 Plotting ACF and PACF...")
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    plot_acf(series, ax=axes[0])
    plot_pacf(series, ax=axes[1])
    plt.suptitle(f"{stock} - ACF & PACF")
    plt.savefig(f"{output_dir}/{stock}_acf_pacf.png")
    plt.close()
    print("✅ ACF & PACF plots saved.\n")

    # Fit ARIMA model (assume p=1, d=1, q=1 as default or based on your ACF/PACF)
    print(f"⚙️ Fitting ARIMA(1,1,1) model for {stock}...")
    model = ARIMA(series, order=(1, 1, 1))
    model_fit = model.fit()

    # Forecast
    forecast = model_fit.predict(start=0, end=len(series)-1, dynamic=False)

    # Evaluate
    mse = mean_squared_error(series[1:], forecast[1:])  # skip first NaN from diff
    print(f"📉 MSE for {stock}: {mse:.4f}")

    # Plot actual vs predicted
    plt.figure(figsize=(10, 5))
    plt.plot(series, label="Actual")
    plt.plot(forecast, label="Predicted")
    plt.title(f"{stock} - Actual vs Predicted")
    plt.legend()
    plt.savefig(f"{output_dir}/{stock}_prediction.png")
    plt.close()
    print("✅ Prediction plot saved.")

    # Save predictions
    forecast_df = pd.DataFrame({'Date': series.index, 'Actual': series.values, 'Predicted': forecast})
    forecast_df.to_csv(f"{output_dir}/{stock}_forecast.csv", index=False)
    print("💾 Forecast saved to CSV.")

    # Optionally save model
    # model_fit.save(f"{output_dir}/{stock}_model.pkl")  # uncomment to save model object

print("\n🎉 All stocks predicted for ARIMA modeling!")

```

# 📈 ARIMA Model Forecasting Report

This notebook performs **ARIMA (AutoRegressive Integrated Moving Average)** time series forecasting on multiple stock datasets. The key steps and results are summarized below:

---

## 📁 Workflow Summary

1. **Data Source**: Reads all CSV files from `../Data/ARIMA/`
2. **Preprocessing**:
   - Parses the `Date` column as the index.
   - Extracts the primary stock price column.
3. **Exploratory Analysis**:
   - Plots **ACF (Autocorrelation Function)** and **PACF (Partial Autocorrelation Function)** for model parameter identification.
4. **Modeling**:
   - Fits an **ARIMA(1,1,1)** model for each stock.
   - Forecasts the stock values across the full date range.
5. **Evaluation**:
   - Computes **Mean Squared Error (MSE)**.
   - Saves actual vs predicted plots for each stock.
6. **Results Export**:
   - Forecasts saved as CSV.
   - Visuals saved as PNG in `results/ARIMA`.
