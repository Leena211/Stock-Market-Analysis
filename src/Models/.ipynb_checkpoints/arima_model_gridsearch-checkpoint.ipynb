```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
warnings.filterwarnings("ignore")
from sklearn.metrics import mean_absolute_error, mean_squared_error
import seaborn as sns

# Custom function for MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Empty list to store metrics for all stocks
evaluation_results = []

# ------------------------------
# CONFIG
# ------------------------------
input_dir = "../Data/ARIMA"
output_dir = "results/ARIMA"
os.makedirs(output_dir, exist_ok=True)

stocks = ['Apple', 'GeneralElectric', 'IBM', 'JohnsonJohnson', 'Microsoft']

# ------------------------------
# FUNCTION: Plot ACF & PACF
# ------------------------------
def plot_acf_pacf(series, stock):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    plot_acf(series, ax=ax[0])
    plot_pacf(series, ax=ax[1])
    ax[0].set_title(f"{stock} ACF")
    ax[1].set_title(f"{stock} PACF")
    plt.tight_layout()
    plt.savefig(f"{output_dir}/{stock}_acf_pacf.png")
    plt.close()

# ------------------------------
# FUNCTION: Evaluate Model
# ------------------------------
def evaluate_model(y_true, y_pred, stock):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    print(f"📉 {stock} RMSE: {rmse:.4f} | MAE: {mae:.4f}")
    return rmse, mae

# ------------------------------
# FUNCTION: Plot Forecast vs Actual
# ------------------------------
def plot_forecast(actual, forecast, stock):
    plt.figure(figsize=(10, 5))
    plt.plot(actual.index, actual, label="Actual", linewidth=2)
    plt.plot(forecast.index, forecast, label="Forecast", linestyle='--')
    plt.title(f"{stock} – Actual vs Forecast")
    plt.legend()
    plt.savefig(f"{output_dir}/{stock}_forecast.png")
    plt.close()

# ------------------------------
# FUNCTION: Grid Search ARIMA
# ------------------------------
def arima_grid_search(series, p_values, d_values, q_values):
    best_score, best_cfg = float("inf"), None
    best_model = None
    for p in p_values:
        for d in d_values:
            for q in q_values:
                try:
                    model = ARIMA(series, order=(p,d,q)).fit()
                    pred = model.predict(start=d, end=len(series)-1)
                    rmse = np.sqrt(mean_squared_error(series[d:], pred))
                    if rmse < best_score:
                        best_score, best_cfg, best_model = rmse, (p,d,q), model
                except:
                    continue
    return best_cfg, best_score, best_model

# ------------------------------
# MAIN LOOP
# ------------------------------
for stock in stocks:
    print(f"\n🔄 Processing {stock}...")

    # Load data
    df = pd.read_csv(f"{input_dir}/{stock}.csv", index_col='Date', parse_dates=True)
    series = df.iloc[:, 0]

    # ACF/PACF plots
    print(f"📊 Plotting ACF and PACF...")
    plot_acf_pacf(series, stock)

    # Grid Search
    print("🔍 Performing Grid Search for best (p,d,q)...")
    p_values = [0, 1]
    d_values = [1]
    q_values = [0, 1]
    best_order, best_rmse, best_model = arima_grid_search(series, p_values, d_values, q_values)
    print(f"✅ Best ARIMA order for {stock}: {best_order} with RMSE: {best_rmse:.4f}")

    # Forecasting
    forecast = best_model.predict(start=best_order[1], end=len(series)-1)
    forecast.index = series.index[best_order[1]:]

    # Evaluation
    rmse, mae = evaluate_model(series[best_order[1]:], forecast, stock)

    # Plotting
    plot_forecast(series, forecast, stock)

    # Save forecast
    out_df = pd.DataFrame({"Actual": series, "Forecast": forecast})
    out_df.to_csv(f"{output_dir}/{stock}_forecast.csv")
    print("💾 Forecast CSV saved.")

print("\n✅ All stocks processed.")
```

```python
import pandas as pd
# Convert list to DataFrame
results_df = pd.DataFrame(evaluation_results)

# Save to CSV
results_df.to_csv("results/ARIMA/arima_evaluation_summary.csv", index=False)

# Plot bar charts
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))
metrics = ["MAE", "RMSE", "MAPE"]
for i, metric in enumerate(metrics):
    plt.subplot(1, 3, i+1)
    sns.barplot(x="Stock", y=metric, data=results_df, palette="pastel")
    plt.xticks(rotation=45)
    plt.title(f"{metric} Comparison")

plt.tight_layout()
plt.savefig("results/ARIMA/arima_metric_comparison.png")
plt.show()
```
