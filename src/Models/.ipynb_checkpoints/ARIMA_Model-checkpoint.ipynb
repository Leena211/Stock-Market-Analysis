```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

# Set output directory
input_dir = "data/ARIMA"
output_dir = "data/ARIMA_Results"
os.makedirs(output_dir, exist_ok=True)
print(f"📁 Output directory '{output_dir}' is ready.\n")

# Stock files (already differenced or stationary)
stock_files = os.listdir(input_dir)

for file in stock_files:
    stock_name = file.replace(".csv", "")
    print(f"\n🔍 Working on {stock_name}...")

    # Load stock data
    df = pd.read_csv(os.path.join(input_dir, file), parse_dates=["Date"], index_col="Date")
    series = df[stock_name]

    # Plot ACF and PACF
    print("📊 Plotting ACF & PACF...")
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    plot_acf(series, ax=axes[0], title=f"{stock_name} ACF")
    plot_pacf(series, ax=axes[1], title=f"{stock_name} PACF")
    plt.tight_layout()
    plt.savefig(f"{output_dir}/{stock_name}_ACF_PACF.png")
    plt.close()

    # Fit ARIMA model (choose simple params, e.g. p=1, d=1, q=1)
    print("⚙️ Fitting ARIMA(1,1,1) model...")
    model = ARIMA(series, order=(1, 1, 1))
    model_fit = model.fit()

    # Forecast (in-sample prediction)
    print("🔮 Forecasting (in-sample)...")
    pred = model_fit.predict(start=0, end=len(series)-1)
    
    # Plot actual vs predicted
    print("📈 Plotting actual vs predicted...")
    plt.figure(figsize=(10, 5))
    plt.plot(series, label="Actual", color="blue")
    plt.plot(pred, label="Predicted", color="red")
    plt.title(f"{stock_name} - Actual vs Predicted")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"{output_dir}/{stock_name}_Forecast.png")
    plt.close()

    # Save predictions
    pred_df = pd.DataFrame({
        "Date": series.index,
        "Actual": series.values,
        "Predicted": pred.values
    })
    pred_df.set_index("Date", inplace=True)
    pred_df.to_csv(f"{output_dir}/{stock_name}_Predictions.csv")

    print(f"✅ Completed for {stock_name}. Results saved.\n")
```
