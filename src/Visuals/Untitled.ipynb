```python
import pandas as pd

# Example result dictionary from model evaluations
results_summary = {
    'Stock': ['Apple', 'GeneralElectric', 'IBM', 'JohnsonJohnson', 'Microsoft'],
    'MAE': [1.2, 1.5, 1.1, 0.9, 1.8],
    'RMSE': [1.5, 1.7, 1.4, 1.0, 2.0],
    'MAPE': [5.2, 6.0, 4.8, 4.1, 6.5]
}

# Convert to DataFrame
results_df = pd.DataFrame(results_summary)
# Melt the DataFrame for seaborn
results_melted = pd.melt(results_df, id_vars="Stock", 
                         value_vars=["MAE", "RMSE", "MAPE"],
                         var_name="Metric", value_name="Score")
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(data=results_melted, x="Stock", y="Score", hue="Metric", palette="pastel")
plt.title("Stock-wise Model Performance Comparison (MAE, RMSE, MAPE)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
